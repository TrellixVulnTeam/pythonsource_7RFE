```python
import pandas as pd
```

```python
temp_df = pd.read_csv("./ta.csv", encoding="ms949")
temp_df
```

```python
temp_df.info()
```

```python
temp_max = pd.DataFrame(temp_df["최고기온(℃)"])
temp_max # 데이터프레임으로 담기
```

```python
# isnull 제거 (원본데이터있음) inplace=True 테이블에 바로 반영

temp_max.dropna(inplace=True)
```

```python
temp_max.head()
```

```python
plt.plot(temp_max["최고기온(℃)"],"red")
plt.show()
```

```python
# 각 년도의 8월 데이터만 추출하여 시각화
```

```python
temp_month = temp_df[["날짜","최고기온(℃)"]].copy()
temp_month                     
```

```python
# 월만 추출하기
def get_month(date):
    return date.split("-")[1] # ['2000','01','01'] 리스트구조로 넘어옴
```

```python
temp_month["월"] = temp_month["날짜"].apply(get_month)
temp_month
```

```python
temp_month.info()
```

```python
temp_month.dropna(inplace=True) #null 데이터 제거후 바로 반영
```

```python
# 8월 데이터

temp_8 = temp_month[temp_month["월"]=="08"]
temp_8
```

```python
temp_8.reset_index(inplace=True) #인덱스 새로부여
temp_8
```

```python
plt.plot(temp_8["최고기온(℃)"],"red")
plt.show()
```

```python
# 각 년도의 오늘 날짜의 최고기온을 시각화

temp_month            
```

```python
# 일만 추출하기
def get_day(date):
    return date.split("-")[2] # ['2000','01','01'] 리스트구조로 넘어옴
```

```python
temp_month["일"] = temp_month["날짜"].apply(get_day)
temp_month
```

```python
temp_today = temp_month[(temp_month["월"]=="03") & (temp_month["일"]=="17")]
temp_today
```

```python
plt.plot(temp_today["최고기온(℃)"],"darkgreen")
plt.show()
```

```python
# 1990년 3/17 이후의 최고기온 시각화

temp_df.info()
```

```python
temp_high = temp_df[["날짜","최고기온(℃)"]].copy()
temp_high
```

```python
temp_high.dropna(inplace=True)
```

```python
temp_high["년도2"] = pd.to_datetime(temp_high["날짜"]) # to_datetime :날짜 형식으로 변환
temp_high.info()
```

```python
temp_high["년도"] = temp_high["년도2"].dt.year
temp_high["월"] = temp_high["년도2"].dt.month
temp_high["일"] = temp_high["년도2"].dt.day
temp_high
```

```python
temp_high.info()
```

```python
temp_today = temp_high[(temp_high["년도"] >= 1990) & (temp_high["월"] == 3) & (temp_high["일"] == 17)]
temp_today
```

```python
plt.plot(temp_today["최고기온(℃)"],"red")
plt.show()
```
