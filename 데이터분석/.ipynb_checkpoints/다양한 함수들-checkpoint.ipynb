```python
df = pd.DataFrame({"a":[4,5,6,6], "b":[7, 8, 9, 9], "c":[10, 11, 12, 12]}, index=['a', 'b', 'c', 'd'])
df
```

```python
df.drop_duplicates() #중복데이터 제거(원본데이터 있음)
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.duplicated() # boolean 타입으로 반환, 중복 여부만 확인
```

```python
df
```

```python
# b 컬럼의 값 중 7이 아닌 것 조회

df["b"] !=7
```

```python
df[df["b"] !=7]
```

```python
# b 컬럼의 값 중 7이 아닌 것 중에서 a, b 컬럼만 조회할 경우

df.loc[df["b"]!=7,["a","b"]]
```

```python
# isin : 특정 컬럼에 해당 값이 들어 있는지 확인할때 사용

df.a.isin([5])

df["a"].isin([5])
```

```python
# isnull() : 널값 확인

df = pd.DataFrame(
    {
        "a":[4,5,6,6, np.nan],
        "b":[7,8,np.nan,9,9],
        "c":[10,11,12,np.nan,12]
    },
    index = [1,2,3,4,5]
)
df
```

```python
df.isnull()
```

```python
df["a"].isnull() #열 지정하여 널 값 확인
```

```python
df.isnull().sum() #컬럼의 널 개수의 합계
```

```python
df["b"].isnull().sum() #b 열의 널 값 합계
```

```python
# notnull() : isnull()의 반대

df.notnull()
```

```python
df["c"].notnull()
```

```python
df.notnull().sum() #notnull 갯수 합계
```

```python
df.a.notnull().sum() #a 컬럼의 notnull 갯수합계
```

```python
4 and 5
```

```python
# pandas 에서는 and 사용 불가 => & 사용

df[df.b == 7] and df[df.a == 5]
```

```python
df[(df.b == 7) & (df.a == 5)]
```

```python
df[(df.b == 7) | (df.a == 5)]
```
