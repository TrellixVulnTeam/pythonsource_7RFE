```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "name":['Alfred','Batman','Catwoman'],
    "toy":[np.nan,"Betmobile",'Bullwhip'],
    "born":[pd.NaT, pd.Timestamp("1940-04-25"), pd.NaT]
})
df
```

```python
df.dropna? #null 값을 어떻게 사용하는지 알고 싶을때 ? 로 조회
```

```python
# axis = 0(행), 1(열) #(원본데이터 유지)
# how=any 행or열에 널값이 하나라도 있으면 삭제

df.dropna(axis=1, how="any") #열 기준 nan처리
```

```python
df.dropna(axis=0, how="any") #행 기준 nan 처리
```

```python
# how=all 널값이 모두 존재해야 삭제

df.dropna(axis=1, how="all")
```

```python
df.fillna? #널 값을 다른 값을 채울때
```

```python
df.fillna(0)
```

```python
values={"name":"noname","toy":"bat","born":pd.Timestamp("1900-01-01")}
df.fillna(value=values)
```

```python
student_list = {
    "name":["Nate", "Yuna", "Abraham", "Brian", "Janny", "Nate", "John"],
    "job": ["teacher","teacher","teacher","student","student","student","teacher"],
    "age":[40,35,37,10,11,None,None]
}
df = pd.DataFrame(student_list)
df
```

```python
df.info() #null이 존재하는지 정보확인
```

```python
df.isnull()
```

```python
df.isna()
```

```python
df.age = df.age.fillna(0)
df
```

```python
# student 나이가 없는 경우 student 중간값으로 대체
# teacher 나이가 없는 경우 teacher 중간값으로 대체

df.groupby("job")["age"].transform("median")
```

```python
df['age'].fillna(df.groupby("job")["age"].transform("median"), inplace=True)
df
```

```python
df = pd.DataFrame([
    [np.nan,2,np.nan,0],
    [3,4,np.nan,1],
    [np.nan, np.nan, np.nan, 5]
], columns = list('ABCD'))
df
```

```python
# 결측치 확인하기

print(df.shape)
df.info()
```

```python
df.notnull()
```

```python
# 모든 결측치 0으로 채우기
df.fillna(0)
```

```python
# 모든 결측지 특정 값으로 채우기 => A:0, B:1, C:2, D:3
values = {"A":0,"B":1,"C":2,"D":3}
df.fillna(value=values)
```

```python
# 모든 결측치 D 열의 중간값으로 채우기
df.fillna(df['D'].median())
```

```python
# 모든 결측지 D 열의 최대값으로 채우기
df.fillna(df['D'].max())
```
